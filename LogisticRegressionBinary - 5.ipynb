```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv("6-bank_customers.csv")
```

```python
print(df.shape)
print(df.columns)
print(df.info())
print(df.isnull().sum())
```

```python
df.head()
```

```python
# Bu veri setinde müşterilerin bazı özellikleri var.
# Bu özelliklere işinden memnun oluşunun sayısal hali vs. gibi özelliklerdir.
# Veri setimizide null değer yok, scaling ve encoding işlemleri yapılmış. Yani gayet temiz bir datasetimiz bulunmakta.
# Bu veri setinde terget kolonumuz subscribed kolonudur.
# Independent feature'lardan yola çıkarak müşterinin abone olup olmadığını tahmin edeceğiz.
# Yani müşterileri abone olup olmama durumuna göre sınıflandıracağız.
```

```python
from sklearn.model_selection import train_test_split
X = df.drop("subscribed", axis=1)
y = df["subscribed"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
# Zaten hazır olan datasetimizi train_test_split ile ayırdık.
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_pred = logistic.predict(X_test)
```

```python
# Veriler zaten scaled olduğu için direkt linear_model'den LogisticRegression'u import ediyoruz.
# Modelimizi eğitiyoruz ve test verilerimizi ile tahminlerimizi yapıyoruz.
```

```python
print(y_pred)
```

```python
# Tahmin sonuçlarımızın hepsi 1 ve 0 şeklinde çünkü bu bir sınıflandırma problemi
# Müşterinin abone olup olmayacağını tahmin etmek ve buna göre sınıflandırmak istiyoruz.
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuarcy = accuracy_score(y_pred, y_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Accuarcy: ", accuarcy)
print("Confusion Matrix: \n", conf_matrix)
print("Classificiaton Report: \n ", classification_report(y_pred, y_test))
```

```python
# sklearn.metrics'ten sınıflandırma başarı metrikleri olan accuracy_score ve confusion_matrix'i içe aktardık.
# Bir de classification_report metodunu içe aktardık. Bu metodla beraber F1 skorunu, Precision'u ve Recall'ı görebileceğiz.
# Modelimizin başarı metriklerini hesaplarken parametre olarak tahmin ettiğimiz veriler ve gerçek verileri veriyoruz.
# Modelimizin başarısı genel olarak gayet yüksek, hem precision hem de recall açısından güçlü ve doğruluğı yüksek.
# Hem de hiçbir hiperparametre ayarlaması yapmadık. Bu dataset nedeniyle oldu.
# Datasetimizi çok temizdi ve hiçbir ayarlama yapmamıza gerek kalmadı.
# Fakat biz bu hiperparametrelere müdahele ederek model başarımlarını optimize edebiliriz.
# Ve bu diğer sınıflandırma algoritmalarında model başarımını arttırmak için çok sık yapılan bir şeydir.
```

## Hyperparameter Tuning

```python
model = LogisticRegression()
```

```python
penalty = ["l1", "l2", "elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01]
solvers = ["lbfgs", "liblinear", "newton-cg", "newton-cholesky", "sag", "saga"]
params = dict(penalty=penalty, C=c_values, solver=solvers)
params
```

```python
# Modelimizin performnasını artırmak için kendimiz ayarlayabileceğimiz parametreler ve hiperparametreler vardır.
# Bunları kendimiz küçük küçük değiştirerek ve sonuçları izleyerek en optimum parametreleri bulabiliriz.
# Ama bunlar için dataset hakkında bilgi sahibi olmamız lazım. Biz daha EDA bile yapmadık.
# Bu nedenle tüm parametreleri bir liste içine kaydettik ve yüm kombinasyonları deneyerek en iyi sonuç veren kombinasyonu deneyeceğiz.
# Bunu normal for-loop ile de yapabiliriz fakat bunu bizim için otomatik olarak yapan GridSearchCV diye bir sınıf var.
# Bu sınıf üzerinden cross-validation işlemine de müdahele edip model başarımını optimize edebiliriz.
```

```python
# Burada 3 parametre ayarlaması yapacağız. Bunlardan ilki penalty ve bu ceza demek.
# Yani hangi regularizasyon tekniğini uygulayacağımızı belirtiyoruz.
# İkincisi ise C değeri ve burada da regularizasyonun gücünü belirliyoruz. 
# Burada parametreleri rastgele verdik çünkü hangi aralığın en iyi sonucu vereceği hakkında bir fikrimiz yok.
# Üçüncüsü ise solvers parametresidir. Burada lojistik regresyonun hangi algoritma ile optimize edileceğini seçeriz
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model, param_grid=params, cv=cv, scoring="accuracy", n_jobs=-1)
grid.fit(X_train, y_train)
```

```python
# Öncelikle model_selection'dan GridSearchCV ce StratifiedKFold sınıflarını getiriyoruz.
# cv diye bir değişkeni StratifiedKFold sınıfndan oluşturuyoruz. Bu değişkeni cross-validation için kullanacağız.
# cv değişkenine default olarak 5 kere kfold cross validation yapması için değer atandı ve biz bunu değiştirebiliriz.
# Daha sonra grid diye bir değişkeni GridSearchCV sınıfından oluşturacağız ve burada bir sürü parametre vereceğiz.
# ilk vereceğimiz parametre olan estimator bizim model değişkeninde bulundan LogisticRegression'dur.
# İkinci parametremiz olan param_grid'e penalty, C ve solver parametrelerini tuttuğumuz sözlüğü veririz.
# Diğer parametreye (cv) Cross-validation objemiz olan cv'yi veririz.
# scoring parametresi kombinasyonları denerken hangş sonuca göre deneyeceğini belirtir.
# Biz burada accuracy skoruna göre en iyi kombinasyonun seçilmesini istiyoruz.
# Daha sonra n_jobs=-1 diyoruz. Böylece modelimizi eğitirken işlemcinin sadece buna odaklanmasını ve daha hızlı olmasını sağlıyoruz.
# Daha sonra parametrelerini ayarladığımız modeli eğitiyoruz.
```

```python
print(grid.best_params_)
print(grid.best_estimator_)
```

```python
# Modelimizde bazı instance'lar vardır ve bunlar en iyi estimator ve en iyi parametre değerlerini tutarlar.
# Mesela bizim için en iyi solver liblinear ve en iyi C değeri 0.1'miş
```

```python
y_pred = grid.predict(X_test)
accuarcy = accuracy_score(y_pred, y_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Accuarcy: ", accuarcy)
print("Confusion Matrix: \n", conf_matrix)
print("Classificiaton Report: \n ", classification_report(y_pred, y_test))
```

```python
# Hiperparametre ayarlamasının ardından sonuçlarımızı inceliyoruz ve sonuçlarımızda küçük de olsa bir artış olduğunu gözlemleyebiliyoruz.
# Modelimizin doğruluğu artmış ve doğru tahmin sayısı 1 artmış. 
# Bu veri seti çok temiz olduğu için hiperparametre ayarlaması sonuçları çok arttırmayabilir.
# Fakat kirli ve gürültülü datasetlerde hiperparametre ayarlaması çok önemlidir. Hatta normal modeli eğitmekten dahi önemlidir.
# Çünkü bu işlemleri matematiksel olarak yapmayı bilmezsek modellerin çalışmasına müdahale edemez ve optimize edemeyiz.
```

## RandomizedCV

```python
from sklearn.model_selection import RandomizedSearchCV
model2 = LogisticRegression()
randomcv = RandomizedSearchCV(estimator=model2, param_distributions=params, cv=5, n_iter=10, n_jobs=-1)
randomcv.fit(X_train, y_train)
```

```python
print(randomcv.best_estimator_)
print(randomcv.best_params_)
print(randomcv.best_score_)
```

```python
# Eğer hyperparameter tuning kısmında çok fazla parametre varsa işlemler uzun sürebilir.
# lojistik regresyon basit bir model olduğu için hesaplamalar çok uzun sürmez.
# Fakat Support Vector Machines gibi algoritmalar zor modeller olduğu için onlarda hiperparametre ayarlaması uzun sürebilir.
# Yani saatlerce sürecek hesaplamalar yapılabilir.
# Bu nedenle bazen rastgele kombinasyonları denemek ve en iyisini seçmek mantıklı olabilir.
# Bu gibi durumlarda RandomizedSearchCV kullanabiliriz.
# Burada yine estimator, param_distributions, cv ve n_jobs parametrelerini veririz.
# bir de n_iter diye bir parametre veririz. Bu paramete kaç tane rastgele kombinnasyon deneyeceğimizi belirtir ve genelde 10 yapılır.
```
